```python

# Assignment:1

# Question : 1

# create two random  3*3  matrix

import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

A=np.random.randint(0,10, size=(3,3))
B=np.random.randint(0,10,size=(3,3))

print("Matrix A :\n", A)
print("Matrix B :\n", B)



# Question : 2

matrix_sum=A+B
matrix_difference=A-B
matrix_mul=np.dot(A,B)

print("Matrix sum :\n", matrix_sum)
print("Matrix difference :\n", matrix_difference)
print("Matrix multiplication :\n", matrix_mul)

# Question : 3
det_A = np.linalg.det(A)
print("\nDeterminant of A:", det_A)

if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("\nInverse of A:\n", inv_A)
else:
    print("\nMatrix A is not invertible (Determinant is 0)")


rank_A = np.linalg.matrix_rank(A)
print("\nRank of A:", rank_A)
```

```python
 # Question : 3


import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns


A = np.random.randint(1, 10, size=(4, 4))
B = np.random.randint(1, 10, size=(4, 4))

matrix_sum = A + B
matrix_sub = A - B
matrix_multiplication = np.dot(A, B)

sns.set_style("whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Matrix addition
axes[0].bar(range(1, matrix_sum.size + 1),
            matrix_sum.flatten(),
            color="b")
axes[0].set_title("Matrix Addition")
axes[0].set_xlabel("Index")
axes[0].set_ylabel("Value")

# Matrix subtraction
axes[1].bar(range(1, matrix_sub.size + 1),
            matrix_sub.flatten(),
            color="g")
axes[1].set_title("Matrix Subtraction")
axes[1].set_xlabel("Index")
axes[1].set_ylabel("Value")

# Matrix multiplication
axes[2].bar(range(1, matrix_multiplication.size + 1),
            matrix_multiplication.flatten(),
            color="r")
axes[2].set_title("Matrix Multiplication")
axes[2].set_xlabel("Index")
axes[2].set_ylabel("Value")

plt.tight_layout()
plt.show()
```
